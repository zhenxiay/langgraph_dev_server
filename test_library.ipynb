```python
from sklearn.datasets import fetch_20newsgroups
```

```python
import pandas as pd
test_data = pd.DataFrame(text_data.data[0:5000],columns=['text'])
```

```python
test_data.to_csv('20newsgroups_sci_med.csv', index=False)
```
