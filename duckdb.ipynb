### Set up DB connection

```python
import duckdb
con = duckdb.connect(database='output.duckdb', read_only=True)
```

### Read Data

```python
con.sql("SELECT * FROM result_sentiment_analysis").show()
```

### Export Data

```python
con.sql("SELECT * FROM result_sentiment_analysis")\
    .df()\
    .to_excel("result_sentiment_analysis.xlsx", index=False)
```

### Close Connection

```python
con.close()
```
